```python
def convert(unit, value):
    if unit == 'm':
        return value * 3.28084
    elif unit == 'n':
        return value / 3.28084
    elif unit == 'o':
        return value * 2.20462
    elif unit == 'p':
        return value / 2.20462
    elif unit == 'q':
        return value * 0.264172
    elif unit == 'r':
        return value / 0.264172
    else: 
        return None

print("M: Meters to Ft\nN: Ft to Meters\nO: Kilograms to pounds\nP: Pounds to kilograms\nQ: liters to Gallons\nR: Gallons to Liters")

choose = input("Enter a alphabets(M - R): ").lower()

value = float(input("Enter value: "))
Output = convert(choose, value)
print(f"Final Converted Value: {Output:.4f}")
```

```python
#Task 2

def calc_sum(values):
    return sum(values)
def calc_avg(values):
    return sum(values) / len(values)
def calc_max(values):
    return max(values)
def calc_min(values):
    return min(values)

def main():
    print("Select an Operation: ")
    print("1. Sum") 
    print("2. Average") 
    print("3. Maximum") 
    print("4. Minimum") 

    try:
        operation = int(input("which operation do you want to do: "))

        if operation not in [1,2,3,4]:
            print("Invalid selection.")
            return
        
        numbers_input = input("Enter list of number: ")
        numbers = list(map(float, numbers_input.split()))
        if operation == 1:
            print("Sum:",calc_sum(numbers))
        elif operation == 2:
            print("Average:", calc_avg(numbers))
        elif operation == 3:
            print("Maximum:", calc_max(numbers))
        elif operation == 4:
            print("Minimum:", calc_min(numbers))

    except ValueError:
        print("Invalid Input")

if __name__ == "__main__":
    main()

```

```python
def extract_alternate(elements):
    return elements[::2]

def get_range(elements, start_idx, end_idx):
    return elements[start_idx:end_idx+1]

def reverse_elements(elements):
    return elements[::-1]

def trim_edges(elements):
    return elements[1:-1]

def get_first_k(elements, k):
    return elements[:k]

def get_last_k(elements, k):
    return elements[-k:]

def reverse_skip_alternate(elements):
    return elements[-2::-2]

data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(extract_alternate(data))
print(get_range(data, 2, 5))
print(reverse_elements(data))
print(trim_edges(data))
print(get_first_k(data, 4))
print(get_last_k(data, 3))
print(reverse_skip_alternate(data))
```

```python
def flatten_list(nested_list):
    flat_list = []
    for element in nested_list:
        if isinstance(element, list):
            flat_list.extend(flatten_list(element))
        else:
            flat_list.append(element)
    return flat_list

def access_element(nested_list, path):
    for idx in path:
        nested_list = nested_list[idx]
    return nested_list

def sum_of_elements(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_of_elements(element)
        else:
            total += element
    return total

def remove_value(nested_list, value):
    for i in range(len(nested_list)):
        if isinstance(nested_list[i], list):
            nested_list[i] = remove_value(nested_list[i], value)
        else:
            if nested_list[i] == value:
                nested_list[i] = None
    return [item for item in nested_list if item is not None]
def find_max_value(nested_list):
    max_value = float('-inf')
    for element in nested_list:
        if isinstance(element, list):
            max_value = max(max_value, find_max_value(element))
        else:
            max_value = max(max_value, element)
    return max_value

def count_value_occurrences(nested_list, value):
    count = 0
    for element in nested_list:
        if isinstance(element, list):
            count += count_value_occurrences(element, value)
        else:
            if element == value:
                count += 1
    return count

def deep_flatten_list(nested_list):
    flat_list = []
    for element in nested_list:
        if isinstance(element, list):
            flat_list.extend(deep_flatten_list(element))
        else:
            flat_list.append(element)
    return flat_list

def average_of_elements(nested_list):
    def calculate_sum_and_count(nested_list):
        total_sum = 0
        count = 0
        for element in nested_list:
            if isinstance(element, list):
                nested_sum, nested_count = calculate_sum_and_count(element)
                total_sum += nested_sum
                count += nested_count
            else:
                total_sum += element
                count += 1
        return total_sum, count

    total_sum, count = calculate_sum_and_count(nested_list)
    
    if count == 0:
        return 0
    else:
        return total_sum / count
    
```

```python
import numpy as np


X = np.array([[3,4], [7,8]])
Y = np.array([[5,3], [2,1]])

#calculating of inverse of x
X_inverse = np.linalg.inv(X)

identity_matrix = np.dot(X, X_inverse)
print("X * X_inverse:\n ", identity_matrix)

#Matrix Multiplication
XY = np.dot(X, Y)
YX = np.dot(Y, X)
print("XY:\n ", XY)
print("YX:\n", YX)

#Transpose verification
XY_Tp = XY.T
YTp_XTp = np.dot(Y.T, X.T)

print("Transpose of XY:\n", XY_Tp)
print("Y^T * X^T:\n", YTp_XTp)

Z = np.array([[2, -3, 1], [1, -1, 2], [3,1, -1]])
A = np.array([[-1], [-3], [9]])

Z_inv = np.linalg.inv(Z)
soln1 = np.dot(Z_inv, A)
print("\n Output using inverse:\n", soln1)

soln2 = np.linalg.solve(Z, A)
print("\n Output using np.linalg.solve:\n", soln2)

```
